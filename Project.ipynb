```python

def quiz():
    questions = [
        ("What is the capital of France?", "Paris"),
        ("What is 5 + 3?", "8"),
        ("Who wrote 'To Kill a Mockingbird'?", "Harper Lee"),
        ("What is the largest planet in our solar system?", "Jupiter"),
        ("What is the boiling point of water in Celsius?", "100"),
        ("What is the chemical symbol for gold?", "Au"),
        ("Who painted the Mona Lisa?", "Leonardo da Vinci"),
        ("What is the smallest prime number?", "2"),
        ("What is the speed of light (in km/s)?", "300000"),
        ("Who developed the theory of relativity?", "Einstein")
    ]

    score = 0
    max_score = len(questions) * 5  # Maximum possible score (10 Ã— 5 = 50)

    print("Welcome to the Quiz!")

    for i, (question, answer) in enumerate(questions):
        user_answer = input(f"Q{i+1}: {question} ").strip()
        if user_answer.lower() == answer.lower():
            score += 5
            print("Correct! +5 points")
        else:
            score -= 2
            print("Incorrect! -2 points")

    print("\nQuiz Completed!")
    print(f"Your total score: {score}")

    # Calculate percentage
    percentage = (score / max_score) * 100

    print(f"Your percentage: {percentage:.2f}%")




# Run the quiz
quiz()
```
